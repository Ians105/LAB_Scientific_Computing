```python
# Newton Raphson

# import libraries
import numpy as np

# step 1: Insert the equations
def f(x):
    return x**6 + 2 * x**2 - 3

def df(x):
    return 6 * x**5 + 4 * x

# step 2: Initialize the variables
x0 = 4 # initial guess
tolerance = 0.01 # tolerance
max_iter = 15 # maximum number of iterations

# step 3: Implement the Newton Raphson method
def newt_Raph(x0, tolerance, iteration):
  x1 = x0 - (f(x0)/df(x0))

  if np.abs(f(x1)) < tolerance:
    print(f"The root of equation is: {x1} at iteration {iteration}")
    return
  
  print(f"Iteration of {iteration}. new root = {x1}")

  if iteration >= 15:
    print("Root now found, max iteration reached")
    return

  newt_Raph(x1, tolerance, iteration+1)

# step 4: Print the root
newt_Raph(x0, tolerance, 1)
```

```python
# Bisection method

# step 1: Insert the equations
def f(x):
    return 10 * x**3 + 8 * x**2 + 9

# step 2: Initialize the variables
coordinates = [
    [3, 4],
    [-10, 17],
    [10, -15],
    [-2, 4],
    [12, -5],
    [-8, -13],
    [7, -21],
]
tolerance = 0.001

# step 3: Implement the Bisection method
def bisection(x0, x1, tolerance):
  if np.sign(f(x0))==np.sign(f(x1)):
    print(f"a = {a} and b = {b} not bound a root")
    return
  
  m=(x0+x1)/2

  if np.abs(f(m)) < tolerance:
    print(f"Root of the function: {m} with coordinate x1 = {a} and x2 = {b}")
    return
  elif np.sign(f(m)) == np.sign(f(x0)):
    bisection(m, x1, tolerance)
  elif np.sign(f(m)) == np.sign(f(x1)):
    bisection(x0, m, tolerance)

# step 4: Print the root
for i in coordinates:
  a, b = i
  bisection(i[0], i[1], tolerance)
```
