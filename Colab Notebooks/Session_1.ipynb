```python
firstNum=int(input("Enter the first number: "))
secNum=int(input("Enter the second number: "))
operation=input("Enter the operation: ")

if(operation=="+" or operation=="plus"):
  print(firstNum+secNum)
elif(operation=="-" or operation=="minus"):
  print(firstNum-secNum)
elif(operation=="*" or operation=="multiply"):
  print(firstNum*secNum)
elif(operation=="/" or operation=="division"):
  print(firstNum/secNum)
elif(operation=="%" or operation=="modulo"):
  print(firstNum%secNum)
elif(operation=="^" or operation=="power"):
  print(firstNum**secNum)
```
