```python
import numpy as np
```

```python
def f(x):
  return x**2-5
```

```python
a=2
b=5
n=5
lebar=(b-a)/(n-1)
x=np.linspace(a, b, n)
y=f(x)

leftRiemann=lebar*sum(y[:n-1])
rightRiemann=lebar*sum(y[1:])

x_mid=(x[:n-1]+x[1:])/2
y_mid=f(x_mid)

midRiemann=lebar*sum(y_mid)

trapezoid=lebar/2*(y[0]+(2*sum(y[1:n-1]))+y[n-1])
trapezoid2=lebar/2*(2*(sum(y))-y[0]-y[n-1])
trapezoid3=(leftRiemann+rightRiemann)/2

print(leftRiemann)
print(rightRiemann)
print(midRiemann)
print(trapezoid)
print(trapezoid2)
print(trapezoid3)
```
