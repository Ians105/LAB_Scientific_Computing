**Root of Equation**

Find x that the f(x) is equal or close to 0



**Newton Raphson Method**

x1 = x0 - (f(x0)/g(x0))

f(x1) close or equal to 0

**Bisection**

```python
import numpy as np

def f(x):
  return x**2-5

def g(x):
  return 2*x

def newton_raphson(x0, tolerance):
  x1 = x0 - (f(x0)/g(x0))
  if np.abs(f(x1)) < tolerance:
    print(f"root is found at x = {x1}")
    return
  newton_raphson(x1, tolerance)

newton_raphson(3, 0.001)

def bisection(x0, x1, tolerance):
  if np.sign(f(x0)) == np.sign(f(x1)):
    print("Invalid interval")
    return
  
  m=(x0+x1)/2

  if np.abs(f(m)) < tolerance:
    print(f"Root is found at x = {m}")
    return
  elif np.sign(f(m)) == np.sign(f(x0)):
    bisection(m, x1, tolerance)
  elif np.sign(f(m)) == np.sign(f(x1)):
    bisection(x0, m, tolerance)
  
bisection(1, 3, 0.001)
```
