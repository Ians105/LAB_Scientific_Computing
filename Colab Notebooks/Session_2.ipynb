```python
# 6a + 3b + 2c = 15
# 2a + 7b + 3c = 6
# 7a + 5b + 15c = 12

# 20x + y - 2z = 17
# 3x + 20y - z = -18
# 2x - 3y + 20z = 25

Xs=[
    [20, 1, -2],
    [3, 20, -1],
    [2, -3, 20]
]

Ys=[17, -18, 25]

import numpy as np
def gauss_seidel(x, y):
  x=np.array(x)
  y=np.array(y)
  conv=0.022
  n=15
  mat_old=np.zeros(np.shape(x[0]))

  # check if the problem is diagonally dominant
  diag=np.array(np.diag(x));
  diag_abs=np.array(np.abs(np.diag(x)))
  np.fill_diagonal(x, 0)
  off_diag=np.array(np.sum(x, axis=1))
  
  if not np.all(diag_abs>off_diag):
    print("The problem is not diagonally dominant")
    return
  
  for i in range(n):
    mat_new=np.array(mat_old)
    for k, row in enumerate(x):
      mat_new[k]=(y[k]-np.dot(row, mat_new))/diag[k]
    print("Iteration number ", i+1, "=", mat_new)
    dx=np.sqrt(np.dot(mat_new-mat_old, mat_new-mat_old))
    mat_old=mat_new
    
    if dx<conv:
      print("The solution is convergent")
      return
    
  print("Not convergent, reached the maximum iteration")
  # print(off_diag)

gauss_seidel(Xs, Ys)
```
